```python
import os
import pandas as pd
import xml.etree.ElementTree as ET
import html
```

```python
# Definir los directorios de entrada y salida
input_dir = '../xmls/xml_files'
output_dir = 'txt_files'
```

```python
# Definir cantidad de archivos a procesar desde ../ids/ids.csv de la ultima columna.

ids_df = pd.read_csv('../ids/ids.csv')
n_files = len(ids_df.iloc[:, -1])
```

```python
# Crear el directorio de salida si no existe
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
```

```python
# Funcion para convertir referencias html a texto plano
def html_to_plain_text(text):
    return html.unescape(text)
```

```python
# Función para limpiar el contenido del texto
def clean(xml):
    text = html_to_plain_text(xml)
    return text
```

```python
# Leer XML y escribirlo en pantalla 
xml_file_path = os.path.join(input_dir, '1.xml')
tree = ET.parse(xml_file_path)
print(ET.tostring(tree.getroot(), encoding='utf8').decode('utf8'))
```
